```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
```

```python
# Carregando o dataset
df = pd.read_csv("../00_dataset/models_results.csv")
```

```python
# Dados estatisticos com o desempenho dos algoritmos
df.describe()
```

```python
# Distribuição com os resultados por Acurácia, Desvio padrão e Tempo de execução
sns.relplot(x=df["Algoritmo"],y=df["Acurácia K-fold"], hue=df["Algoritmo"], size=df["Algoritmo"], data=df);
sns.relplot(x=df["Algoritmo"],y=df["Desvio padrão"], hue=df["Algoritmo"], size=df["Algoritmo"], data=df);
sns.relplot(x=df["Algoritmo"],y=df["Tempo de execução (s)"], hue=df["Algoritmo"], size=df["Algoritmo"], data=df);
```

```python
# Algoritmo que apresentou a melhor acurácia
df.loc[df['Acurácia K-fold'].idxmax()]
```

```python
# Algoritmo que apresentou o menor desvio padrão
df.loc[df['Desvio padrão'].idxmin()]
```

```python
# Algoritmo que apresentou o menor custo computacional para execução
df.loc[df['Tempo de execução (s)'].idxmin()]
```
